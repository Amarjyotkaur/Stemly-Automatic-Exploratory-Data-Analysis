```python
import features as ft
import pandas as pd
```

# Column Data 
as Pandas Series

```python
a=pd.Series([1,2,3,4,5,4,3,2,2,1])  #discrete categorical numeric
b=pd.Series([1.25,3.44,5.55,1.23,8.43,9.76])  #Continuous numerical
c = pd.Series([20.5,23.6,30.8,27.2,28.3,22.1,25.7,0,2.1,27.3,None,100.2])  #continuous with presence of outliers, Zero and Missing Value.
d =pd.Series(["cat","monkeys", "a","dog","monkey","dog","cat","dog"])#categorical nominal
data =[a,b,c,d]
```

```python
lsSize=[]
lsNullvalues=[]
lsNullratio=[]
lsDatatype=[]
lsMemory=[]
lsDistinctcount=[]
lsCardinality=[]
lsCount=[]
lsCountbin=[]
lsMin=[]
lsMax=[]
lsMean=[]
lsMedian=[]
lsVariance= []
lsSum=[]
lsType=[]
lsStddev=[]
lsUniqvalues=[]
lsFrequency=[]
lsMode=[]
lsRange=[]
lsQuantile25=[]
lsQuantile75=[]
lsQuantile10=[]
lsQuantile90=[]
lsQuantile1=[]
lsQuantile99=[]
lsIqr=[]
lsKurtosis=[]
lsSkewness=[]
lsMeanabsdev=[]
lsOutlier=[]
lsNumoutliers=[]
lsZerocount=[]
lsCoefofvar=[]
lsMinf=[]
lsMaxf=[]
lsMeanf=[]
lsMedianf=[]
lsVariancef=[]
lsCoefofvarf=[]
lsStddevf=[]
lsRangef=[]
lsMeanabsdevf=[]
lsOutlierf=[]
lsNumoutliersf=[]
lsMostleastfreq=[]
lsMissingintc=[]
lsMissingintr=[]
lsClass=[]


```

# General Info from Column Data 

Size of Column,
Number of Missing Values,
Ratio of Missing Values,
Data Type,
Attribute Type,
Memory Size

```python
for i in data:
        print("number of null/missing vales: ",ft.null_c(i))
        lsNullvalues.append(ft.null_c(i))
        print("Ratio of null values: ", ft.null_r(i))
        lsNullratio.append(ft.null_r(i))
        print("Size: ", ft.size(i))
        lsSize.append(ft.size(i))
        print("Memory Size: ", ft.memory(i), "Bytes")
        lsMemory.append(ft.memory(i))
        print("Type: ", ft.type(i))
        lsType.append(ft.type(i))
        print("Attribute Class: ",ft.att_class(i))
        lsClass.append(ft.att_class(i))
        print("Missing Integer Count: ", ft.missing_int_c(i))
        lsMissingintc.append(ft.missing_int_c(i))
        print("Missing Integer Ratio: ", ft.missing_int_r(i),"\n")
        lsMissingintr.append(ft.missing_int_r(i))
        
        
```

# Measures of Extremes 
 Minimum, Maximum, Number of Outliers, List of Outlier Values, Count of Zero Values

```python
#Minimum and Maximum

for i in data:
    try:
        print("minimum is ",ft.min(i))
        lsMin.append(ft.min(i))
        print("maximum is ",ft.max(i))
        lsMax.append(max(i))
        print("Number of Outliers: ", ft.outlier_c(i))
        lsNumoutliers.append(ft.outlier_c(i))
        print("Outliers: ", ft.outlier(i))
        lsOutlier.append(ft.outlier(i))
        print("Number of Zeros: ", ft.zero_c(i),"\n")
        lsZerocount.append(ft.zero_c(i))
    except:
        print('feature not available for Categorical Data')

       
```

# Measures of Central Tendency
Mean, Median

```python
# MEAN And MEDIAN
#pd.mean(pd.Series(a))
for i in data:
    try:
        print("mean is ", ft.mean(i))
        lsMean.append(ft.mean(i))
        print("median is ", ft.median(i), "\n")
        lsMedian.append(ft.median(i))
    except:
        print( 'feature not available for Categorical Data')

```

# Measures of Uniqueness
Distinct Count, List of Unique Values, Cardinality Ratio

```python
for i in data:
    print("distinct count:", ft.cardinality_c(i))
    lsDistinctcount.append(ft.cardinality_c(i))
    print("unique values: ", ft.uniq_values(i))
    lsUniqvalues.append(ft.uniq_values(i))
    print("cardinality ratio: ", ft.cardinality_r(i),"\n")
    lsCardinality.append(ft.cardinality_r(i))
```

# Measures of Dispersion 
Variance, Coefficient of Variance, Standard Deviation, Range, Mean Absolute Deviation

```python
for i in data:
    try:
        print("variance is ", ft.variance(i))
        lsVariance.append(ft.variance(i))
        print("Standard Deviation is ", ft.std_dev(i))
        lsStddev.append(ft.std_dev(i))
        print("Range is ", ft.range(i))
        lsRange.append(ft.range(i))
        print("Mean Absolute Deviation: ", ft.mean_abs_dev(i))
        lsMeanabsdev.append(ft.mean_abs_dev(i))
        print("Coefficient of Variance: ",ft.coef_of_var(i))
        lsCoefofvar.append(ft.coef_of_var(i))
        print("\n")
    except:
        print('feature not available for Categorical Data')
```

# Measures of Frequency
Count, Frequency, Mode

```python
for i in data:
  
    print("Count:\n",ft.count_c(i))
    lsCount.append(ft.count_c(i))
    print("frequency %:\n",ft.count_r(i))
    lsFrequency.append(ft.count_r(i))
    print("mode: ", ft.mode(i).iloc[0])
    lsMode.append(ft.mode(i).iloc[0])
    print(ft.most_least_frequent(i))
    lsMostleastfreq.append(ft.most_least_frequent(i))
    print("\n")
    
```

```python
for i in data:
    print("Bins Count: ","\n", ft.count_bins(i),"\n")
    lsCountbin.append(ft.count_bins(i))
```

# Sum

```python
for i in data:
    print( "sum: ", ft.sum(i))
    lsSum.append(ft.sum(i))

    

    
```

# Measures of Position
Quantile Stats, Inter Quartile Range (IQR)

```python
#Quantiles & Deciles
for i in data:
    try:
        print("1st quartile: ", ft.quantile(i,0.25))
        lsQuantile25.append(ft.quantile(i,0.25))
        print("2nd quartile: ", ft.quantile(i,0.5))
        print("3rd quartile: ",ft.quantile(i,0.75))
        lsQuantile75.append(ft.quantile(i,0.75))
        print("4th quartile: ", ft.quantile(i,1))
        print("1st Percentile: ", ft.quantile(i,0.01))
        lsQuantile1.append(ft.quantile(i,0.01))
        print("10th Percentile: ", ft.quantile(i,0.1))
        lsQuantile10.append(ft.quantile(i,0.1))
        print("90th Percentile: ", ft.quantile(i,0.9))
        lsQuantile90.append(ft.quantile(i,0.9))
        print("99th Percentile: ", ft.quantile(i,0.99))
        lsQuantile99.append(ft.quantile(i,0.99))
        print("IQR: ", ft.iqr(i))
        lsIqr.append(ft.iqr(i))
        print("\n")
    except:
        print('feature not available for Categorical Data')
            
    
```

# Measures of Normality/Shape
Kurtosis, Skewness

```python
for i in data:
    try:
        print( "kurtosis: ", ft.kurtosis(i))
        lsKurtosis.append(ft.kurtosis(i))
        print("skewness: ", ft.skewness(i))
        lsSkewness.append(ft.skewness(i))
        print("\n")
    except:
        print('feature not available for Categorical Data')
    
```

# Frequency Based Metrics
Calculated on the frequency distribution of input Column Data

Min, Max, mean, median, variance, standard deviation, range, mean absolute deviation, outliers, number of outliers

```python
for i in data:
    print("Min: ", ft.min_freq(i))
    lsMinf.append(ft.min_freq(i))
    print("Max: ",ft.max_freq(i))
    lsMaxf.append(ft.max_freq(i))
    print("Median: ",ft.median_freq(i))
    lsMedianf.append(ft.median_freq(i))
    print("Mean: ",ft.mean_freq(i))
    lsMeanf.append(ft.mean_freq(i))
    print("Variance ", ft.variance_freq(i))
    lsVariancef.append(ft.variance_freq(i))
    print("Standard Deviation: ", ft.stdev_freq(i))
    lsStddevf.append(ft.stdev_freq(i))
    print("Coefficient of Variance: ", ft.coef_of_var_freq(i))
    lsCoefofvarf.append(ft.coef_of_var_freq(i))
    print("Range: ", ft.range_freq(i))
    lsRangef.append(ft.range_freq(i))
    print("Mean Abs Deviation: ", ft.mean_abs_dev_freq(i))
    lsMeanabsdevf.append(ft.mean_abs_dev_freq(i))
    print("Number of Outliers: ",ft.outlier_c_freq(i))
    lsNumoutliersf.append(ft.outlier_c_freq(i))
    print("Outliers: ", ft.outliers_freq(i),"\n")
    lsOutlierf.append(ft.outliers_freq(i))
```
